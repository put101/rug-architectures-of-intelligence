```python
import matplotlib.pyplot as plt
original_experiment = [
    0.6,
    0.65,
    0.7,
    0.86,
    1.12,
    1.5,
    1.79,
    2.13,
    2.15,
    2.58,
]

model_experiment = [
    0.54,
    0.72,
    0.9,
    1.09,
    1.27,
    1.46,
    1.64,
    1.83,
    2.02,
    2.20,
]

X_iterations = range(1, 11)
# number of iterations are positive numbers from 1 to 10. 
# plot the x axis as the number of iterations

fig, ax = plt.subplots()

ax.plot(X_iterations, original_experiment, label="Original experiment")
ax.plot(X_iterations, model_experiment, label="Model experiment")
ax.set_xlabel("Number of iterations")
ax.set_ylabel("Reaction time seconds")
ax.set_title("Reaction time over iterations")
ax.legend()
ax.set_xticks(X_iterations)
```
